```python
import DataLoader
from matplotlib import pyplot as plt
```

```python
path = 'data_2015/2015-06'
data = DataLoader.load_from_folder(path)
data
```

```python
test_projection = data[['Date', 'Time', 'NumOfClients', 'TransmittedFrames', 'Failed', 'FcsErrors', 'AckFailures']]
test_projection
```

```python
grouped_by_date = test_projection.groupby('Date', as_index=False).sum()
grouped_by_date
```

```python
days = grouped_by_date['Date']
days
```

```python
plt.title('Clients')
plt.plot(days, grouped_by_date['NumOfClients'])
plt.show()
```

```python
plt.title('Frames')
plt.plot(days, grouped_by_date['TransmittedFrames'])
plt.show()
```

```python
plt.title('Failed')
plt.plot(days, grouped_by_date['Failed'])
plt.show()
```

```python
plt.title('FcsErrors')
plt.plot(days, grouped_by_date['FcsErrors'])
plt.show()
```

```python
plt.title('AckFailures')
plt.plot(days, grouped_by_date['AckFailures'])
plt.show()
```
