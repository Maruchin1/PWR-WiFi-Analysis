```python
import DataLoader
from matplotlib import pyplot as plt
```

```python
path = 'data_2015/2015-06'
data = DataLoader.load_from_folder(path)
data
```

```python
test_projection = data[['Date', 'Time', 'NumOfClients', 'TransmittedFrames', 'PoorSnrClients', 'Retries', 'MultipleRetries', 'FrameDuplicates']]
test_projection
```

```python
grouped_by_day = test_projection.groupby('Date', as_index=False).mean()
grouped_by_day
```

```python
days = grouped_by_day['Date']
```

```python
plt.title('Clients')
plt.plot(days, grouped_by_day['NumOfClients'])
plt.show()
```

```python
plt.title('Frames')
plt.plot(days, grouped_by_day['TransmittedFrames'])
plt.plot(days, grouped_by_day['Retries'])
plt.plot(days, grouped_by_day['MultipleRetries'])
plt.show()
```

```python
plt.title('PoorSnrClients')
plt.plot(days, grouped_by_day['PoorSnrClients'])
plt.show()
```

```python
plt.title('Retries')
plt.plot(days, grouped_by_day['Retries'])
plt.show()
```
