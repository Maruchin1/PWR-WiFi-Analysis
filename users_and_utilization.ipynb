```python
import pandas as pd
import DataLoader
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
path = 'data_2015/2015-06'
data = DataLoader.load_from_folder(path)
data
```

```python
test_projection = data[['Date', 'NumOfUsers', 'ChannelUtilization', 'TxUtilization', 'RxUtilization']]
test_projection
```

```python
date_time = test_projection['Date'].str.split('--', expand=True)
test_projection.drop(columns=['Date'])
test_projection[['Date', 'Time']] = date_time
test_projection
```

```python
ordered = test_projection[['Date', 'Time', 'NumOfUsers', 'ChannelUtilization', 'TxUtilization', 'RxUtilization']]
ordered
```

```python
grouped_by_date = ordered.groupby('Date', as_index=False).mean()
grouped_by_date
```

```python
dates = grouped_by_date['Date']
users = grouped_by_date['NumOfUsers']
channelUtilization = grouped_by_date['ChannelUtilization']
tx_utilization = grouped_by_date['TxUtilization']
rx_utilization = grouped_by_date['RxUtilization']
```

```python
plt.title('Users')
plt.plot(dates, users)
plt.show()
```

```python
plt.title('ChannelUtilization')
plt.plot(dates, channelUtilization)
plt.show()
```

```python
plt.title('TxUtilization')
plt.plot(dates, tx_utilization)
plt.show()
```

```python
plt.title('RxUtilization')
plt.plot(dates, rx_utilization)
plt.show()
```
