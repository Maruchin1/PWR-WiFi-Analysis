```python
import DataLoader
from matplotlib import pyplot as plt
```

```python
path = '2015_all.csv'
raw_data = DataLoader.load_single_file(path)
raw_data
```

```python
grouped_by_hour = raw_data.groupby(['Day', 'Hour'], as_index=False).sum()
grouped_by_hour
```

```python
clients_and_problems = grouped_by_hour[['NumOfClients', 'PoorSnrClients', 'Retries', 'MultipleRetries', 'FrameDuplicates']]
clients_and_problems
```

```python
poor_clients = clients_and_problems.sort_values(by=['NumOfClients'])
poor_clients
```

```python
plt.title('Liczba klientów ze słabym połączeniem')
plt.plot(poor_clients['NumOfClients'], poor_clients['PoorSnrClients'])
plt.xlabel('Liczba klientów')
plt.show()
```

```python
percentage_poor = poor_clients['PoorSnrClients'] / poor_clients['NumOfClients']
percentage_poor
```

```python
plt.title('Procent klientów ze słabym połączeniem')
plt.plot(poor_clients['NumOfClients'], percentage_poor)
plt.xlabel('Liczba klientów')
plt.show()

```
