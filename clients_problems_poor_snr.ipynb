```python
import CleanDataLoader
from matplotlib import pyplot as plt
```

```python
path = '2015_all.csv'
raw_data = CleanDataLoader.loadFile(path)
raw_data
```

```python
grouped_by_day = raw_data.groupby(['Day'], as_index=False).sum()
grouped_by_day
```

```python
clients_and_problems = grouped_by_day[['NumOfClients', 'PoorSnrClients']]
clients_and_problems
```

```python
poor_clients = clients_and_problems.sort_values(by=['NumOfClients'])
poor_clients
```

```python
num_of_clients = poor_clients[['NumOfClients']]
thousands = (num_of_clients / 20000).astype(int)
rounded_num_of_clients = thousands * 20000
poor_clients[['NumOfClients']] = rounded_num_of_clients
poor_clients
```

```python
mean_by_thousands = poor_clients.groupby(['NumOfClients'], as_index=False).mean()
mean_by_thousands
```

```python
std_by_thousands = poor_clients.groupby(['NumOfClients'], as_index=False).std()
std_by_thousands
```

```python
plt.title('WpÅ‚yw liczby klientÃ³w korzystajÄ…cych z sieci na liczbÄ™ klientÃ³w ze sÅ‚abym poÅ‚Ä…czeniem')
plt.plot(mean_by_thousands['NumOfClients'], mean_by_thousands['PoorSnrClients'])
plt.fill_between(
    mean_by_thousands['NumOfClients'], 
    mean_by_thousands['PoorSnrClients'] + std_by_thousands['PoorSnrClients'],
    mean_by_thousands['PoorSnrClients'] - std_by_thousands['PoorSnrClients'],
    facecolor='lightblue'
)
plt.xlabel('Liczba klientÃ³w')
plt.ylabel('Liczba klientÃ³w ze sÅ‚abym zasiegiem')
plt.show()
```

```python
percentage_poor = mean_by_thousands['PoorSnrClients'] / mean_by_thousands['NumOfClients']
percentage_poor
```

```python
std_percentage_poor = std_by_thousands['PoorSnrClients'] / mean_by_thousands['NumOfClients']
std_percentage_poor
```

```python
plt.title('WpÅ‚yw liczby klientÃ³w korzystajÄ…cych z sieci na odsetek klientÃ³w ze sÅ‚abym zasiÄ™giem')
plt.plot(mean_by_thousands['NumOfClients'], percentage_poor)
plt.fill_between(
    mean_by_thousands['NumOfClients'],
    percentage_poor + std_percentage_poor,
    percentage_poor - std_percentage_poor,
    facecolor='lightblue'
)
plt.xlabel('Liczba klientÃ³w')
plt.ylabel('Procent klientÃ³w ze sÅ‚abym zasiegiem')
plt.show()

```
