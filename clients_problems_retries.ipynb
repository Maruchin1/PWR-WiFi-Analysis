```python
import CleanDataLoader
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = '2015_all.csv'
raw_data = CleanDataLoader.loadFile(path)
raw_data
```

```python
grouped_by_day = raw_data.groupby(['Day'], as_index=False).sum()
grouped_by_day
```

```python
retries = grouped_by_day[['NumOfClients', 'TransmittedFrames', 'Retries', 'MultipleRetries']].sort_values(by=['NumOfClients'])
retries
```

```python
num_of_clients = retries['NumOfClients']
thousands = (num_of_clients / 20000).astype(int)
rounded_num_of_clients = thousands * 20000
retries['NumOfClients'] = rounded_num_of_clients
retries
```

```python
mean_retries = retries.groupby(['NumOfClients'], as_index=False).mean()
mean_retries
```

```python
x_clients = mean_retries['NumOfClients']
y_retries = mean_retries['Retries']
y_multiple_retries = mean_retries['MultipleRetries']

plt.fill_between(x_clients, y_retries)
plt.fill_between(x_clients, y_multiple_retries)
```

```python
frames = mean_retries['TransmittedFrames']
percent_retries = (y_retries / frames) * 100
percent_multiple_retries = (y_multiple_retries / frames) * 100

mean_percent_retries = np.mean(percent_retries)
std_percent_retries = np.std(percent_retries)

mean_percent_multiple_retries = np.mean(percent_multiple_retries)
std_percent_multiple_retries = np.std(percent_multiple_retries)

bars = np.arange(2)
heights = [mean_percent_retries, mean_percent_multiple_retries]
errors = [std_percent_retries, std_percent_multiple_retries]

plt.title('Procentowy udział problemów wśród przesyłanych klatek')
plt.ylabel('Procent przesyłanych klatek')
plt.xlabel('Problemy')
plt.bar(bars, heights, yerr=errors, color=('royalblue', 'darkorange'), capsize=15)
plt.xticks(bars, ['Retries', 'MultipleRetries'])
```
