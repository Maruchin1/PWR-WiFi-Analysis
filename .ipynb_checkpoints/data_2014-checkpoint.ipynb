```python
import pandas as pd
from matplotlib import pyplot as plt
import DataLoader as dl
```

```python
one_month_path = 'data_2014/2014-04'
one_month = dl.load_from_folder(one_month_path)
one_month
```

```python
date_users_clients = one_month[['Date', 'NumOfUsers', 'NumOfClients']]
date_users_clients
```

```python
grouped_by_date = date_users_clients.groupby('Date', as_index=False).mean()
grouped_by_date
```

```python
dates = grouped_by_date['Date']
num_of_users = grouped_by_date['NumOfUsers']
num_of_clients = grouped_by_date['NumOfClients']

plt.plot(dates, num_of_users)
# plt.plot(dates, num_of_clients)
plt.show()
```
