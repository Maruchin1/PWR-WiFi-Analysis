```python
import CleanDataLoader
from matplotlib import pyplot as plt
```

```python
path = '2015_all.csv'
raw_data = CleanDataLoader.loadFile(path)
raw_data
```

```python
grouped_by_day = raw_data.groupby(['Day'], as_index=False).sum()
grouped_by_day
```

```python
retries = grouped_by_day[['NumOfClients', 'Retries', 'MultipleRetries']].sort_values(by=['NumOfClients'])
retries
```

```python
num_of_clients = retries[['NumOfClients']]
thousands = (num_of_clients / 20000).astype(int)
rounded_num_of_clients = thousands * 20000
retries[['NumOfClients']] = rounded_num_of_clients
retries
```

```python
mean_by_thousands = retries.groupby(['NumOfClients'], as_index=False).mean()
mean_by_thousands
```

```python
std_by_thousands = retries.groupby(['NumOfClients'], as_index=False).std()
std_by_thousands
```

```python
clients = mean_by_thousands['NumOfClients']
retries = mean_by_thousands['Retries']
std_retries = std_by_thousands['Retries']
multiple_retries = mean_by_thousands['MultipleRetries']
std_multiple_retries = std_by_thousands['MultipleRetries']

plt.plot(clients, retries)
plt.plot(clients, multiple_retries)
plt.fill_between(
    clients,
    retries + std_retries,
    retries - std_retries,
    facecolor='lightblue'
)
plt.fill_between(
    clients,
    multiple_retries + std_multiple_retries,
    multiple_retries - std_multiple_retries,
    facecolor='moccasin'
)
```

```python
percentage_retries = mean_by_thousands['Retries'] / mean_by_thousands['NumOfClients']
percentage_multiple_retries = mean_by_thousands['MultipleRetries'] / mean_by_thousands['NumOfClients']
std_percentage_retries = std_retries / clients
std_percentage_multiple_retries = std_multiple_retries / clients
```

```python
plt.plot(clients, percentage_retries)
plt.plot(clients, percentage_multiple_retries)
plt.fill_between(
    clients,
    percentage_retries + std_percentage_retries,
    percentage_retries - std_percentage_retries,
    facecolor='lightblue'
)
plt.fill_between(
    clients,
    percentage_multiple_retries + std_percentage_multiple_retries,
    percentage_multiple_retries - std_percentage_multiple_retries,
    facecolor='moccasin'
)
```
