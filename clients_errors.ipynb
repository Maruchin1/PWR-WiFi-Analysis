```python
path = 'data_2015/2015-11'
raw_data = CleanDataLoader.loadFolder(path)
raw_data
```

```python
grouped_by_hour = raw_data.groupby(['Day', 'Hour'], as_index=False).sum()
grouped_by_hour
```

```python
users_and_errors = grouped_by_hour[['NumOfClients', 'Failed', 'FcsErrors', 'AckFailures']]
users_and_errors
```

```python
ordered_by_clients = users_and_errors.sort_values(by=['NumOfClients'])
ordered_by_clients
```
